```python
!pip install control
import control
import numpy as np
import matplotlib.pyplot as plt

## graficos
## 1. funcoes 
## 2. plotar graficos plt.plot()
## 3. títulos, grades
## 4. limites dos eixos
## 5. mais de uma curva no mesmo gráfico
## 6. legenda plt.legend()
## 7. exportar a figura plt.
## 8. figuras com multiplos paineis plt.subplots()

##############################################

b = -5.0

## gerando array de tempo
t = np.linspace(0,0.2,1400)
## funcao seno
y1 = np.sin(367.0*t) 

## senoide amortecida
# y2 = np.sin(367*t) * np.exp(-b*t) 
y2 = y1 * np.exp(b*t) 


## cria janela
fig1 = plt.figure() ## figure(1)
## plota o grafico
plt.plot(t,y1,label='seno')
plt.plot(t,y2,label='seno amort.')

## coloca a grade
plt.grid()
## coloca a legenda
plt.legend()
## coloca titulo no eixo X
plt.xlabel('tempo (s)')
## coloca titulo no eixo Y
plt.ylabel('y(t)')
## titulo no grafico
plt.title('grafico exemplo')

plt.ylim([-1.5,1.5])

#####################################################

## cria janela
fig2 = plt.figure() ## figure(2)

## y = s0+vo*t+a/2*t**2
y3 = 1 + 3*t + 5*t**2 ## t^2

plt.plot(t,y3)
## coloca a grade
plt.grid()
## coloca a legenda
# plt.legend()
## coloca titulo no eixo X
plt.xlabel('tempo (s)')
## coloca titulo no eixo Y
plt.ylabel('S(t)')
## titulo no grafico
plt.title('Sorvetao')

## tempo importatne
ti = 0.12
## linha no tempo importante
plt.plot([ti,ti],[1,1.8],color='red')

## google!
index = (np.abs(t-ti)).argmin()
## plot do ponto importante
plt.plot(t[index],y3[index],'*k',markersize=12)

##############################################

fig3,ax = plt.subplots(2,1)

y4 = y2 + y3

ax[0].plot(t,y1)
ax[1].plot(t,y2)
# ax[2].plot(t,y3)
# ax[3].plot(t,y4)

ax[0].set_xlabel('t (s)')
ax[1].set_xlabel('t (s)')
ax[0].set_ylabel('y1 (s)')
ax[1].set_ylabel('y2 (s)')

val = 1.2
ax[0].set_ylim([-val,val])
ax[1].set_ylim([-val,val])


##############################################
## COM 4 SUBPLOTS, COMO PROMETIDO
#############################################

fig4,ax = plt.subplots(2,2)

y4 = y2 + y3

ax[0][0].plot(t,y1)
ax[0][1].plot(t,y2)
ax[1][0].plot(t,y3)
ax[1][1].plot(t,y4)

### E UM TRUQUE COM FOR LOOP DE BONUS:
for i in range(2):
    for j in range(2):
        ax[i][j].set_xlabel('t (s)')
        ax[i][j].set_ylabel('y'+str(j+2*i+1)+' (s)')
    

############### plt.show() ## 


```
