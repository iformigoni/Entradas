```python
!pip install control
import control
import numpy as np
import matplotlib.pyplot as plt

## 1a ordem

num = [10]
den = [1,6]

G = control.tf(num,den)

print(G)

p = G.pole() ## pole(G)
z = G.zero() ## zero(G) 

print('polos = ',p)
print('zeros = ',z)

t,y = control.step_response(G) ## [y,t] = step(G)

fig = plt.figure()
plt.plot(t,y)
plt.xlabel('t (s)')
plt.ylabel('y(t)')
plt.grid()
plt.title('Resposta ao degrau: 1a ordem')

info = control.step_info(G)
# {'RiseTime': 0.3721349645242902,
#   'SettlingTime': 0.6628654055588918,
#   'SettlingMin': 1.5112327755194601,
#   'SettlingMax': 1.6649999999999943,
#   'Overshoot': 0,
#   'Undershoot': 0,
#   'Peak': 1.6649999999999943,
#   'PeakTime': 1.1512925464970227,
#   'SteadyStateValue': 1.6666666666666665}

print(info)

print('Ts = ',info['SettlingTime'])
print('up = ',info['Overshoot'])
print('Tr = ',info['RiseTime'])

t,yi = control.impulse_response(G) ## [y,t] = impulse(G)

fig = plt.figure()
plt.plot(t,yi)
plt.xlabel('t (s)')
plt.ylabel('y(t)')
plt.grid()
plt.title('Resposta ao impulso: 1a ordem')

######################################

###2a ordem

G1 = control.tf([9],[1,9,9])
G2 = control.tf([9],[1,2,9])
G3 = control.tf([9],[1,0,9])
G4 = control.tf([9],[1,6,9])

G = [G1,G2,G3,G4]

t = np.linspace(0,7,100)

fig,ax = plt.subplots(1,4) ## figure(1)
labels = ['super-amortecido','sub-amortecido'
          ,'não-amortecido','criticamente amortecido']
for i in range(len(G)):
    print(G[i])
    print(G[i].pole())
    t,y = control.step_response(G[i],t)
    ax[i].plot(t,y)
    ax[i].grid()
    ax[i].set_title(labels[i])
    


 
```
